# üîπ Feature Engineering

```python
# 1. Daily Return (% change)
df['Return'] = df['Close'].pct_change()

# 2. Moving Averages (trend indicators)
df['MA_10'] = df['Close'].rolling(window=10).mean()
df['MA_20'] = df['Close'].rolling(window=20).mean()
df['MA_50'] = df['Close'].rolling(window=50).mean()
df['MA_100'] = df['Close'].rolling(window=100).mean()

# 3. Rolling Volatility (risk measure)
df['Volatility_20'] = df['Return'].rolling(window=20).std()

# 4. Lag Features (previous day values)
df['Lag_1'] = df['Close'].shift(1)   # Yesterday's close
df['Lag_2'] = df['Close'].shift(2)   # Day before yesterday
df['Lag_5'] = df['Close'].shift(5)   # One week ago

# Drop NA (caused by rolling calculations at start)
df = df.dropna()

# Preview engineered features
df.head()
```

```python
#Explanation of Features

#Return ‚Üí shows daily price change (up or down).

#Moving Averages (MA_10, MA_20, MA_50, MA_100) ‚Üí smooth out short-term noise, show trends over time.

#Volatility_20 ‚Üí measures risk by showing how much prices fluctuate.

#Lag Features (Lag_1, Lag_2, Lag_5) ‚Üí let the model see past prices to predict the future.

#Why we use them: Raw prices alone aren‚Äôt enough; adding trends, risk, past values, and returns helps ML models understand stock patterns better.
```

```python
# Load cleaned dataset
df = pd.read_csv("AAPL_cleaned.csv", parse_dates=['Date'], index_col='Date')
```

# Baseline Model (Naive Forecast)

```python

# Predict tomorrow's price as today's price
y_pred_naive = test.shift(1).dropna()
rmse_naive = sqrt(mean_squared_error(test.iloc[1:], y_pred_naive))
mae_naive = mean_absolute_error(test.iloc[1:], y_pred_naive)

print("Naive Forecast RMSE:", rmse_naive)
print("Naive Forecast MAE:", mae_naive)
```

Naive forecast Used yesterday‚Äôs closing price as today‚Äôs prediction.

output: RMSE ‚âà 3, MAE ‚âà low ‚Üí surprisingly strong baseline.

```python
plt.figure(figsize=(10,5))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(y_pred_naive, label='Naive Forecast')
plt.legend()
plt.title("Naive Forecast vs Actual")
plt.show()
```

```python
# Simple ARIMA(5,1,0)
model = ARIMA(train['Close'], order=(5,1,0))
model_fit = model.fit()
forecast = model_fit.forecast(steps=len(test))

rmse_arima = sqrt(mean_squared_error(test, forecast))
mae_arima = mean_absolute_error(test, forecast)

print("ARIMA RMSE:", rmse_arima)
print("ARIMA MAE:", mae_arima)

plt.figure(figsize=(10,5))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(test.index, forecast, label='ARIMA Forecast')
plt.legend()
plt.title("ARIMA Forecast vs Actual")
plt.show()
```

# SARIMA Model {Statistical Models}

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
# ‚úÖ SARIMA Model
# (p, d, q) √ó (P, D, Q, s)
# Here s = 12 for monthly seasonality, adjust based on your data frequency
sarima_model = SARIMAX(train['Close'],
                       order=(1, 1, 1),
                       seasonal_order=(1, 1, 1, 12),
                       enforce_stationarity=False,
                       enforce_invertibility=False)

sarima_result = sarima_model.fit(disp=False)

```

```python
# ‚úÖ Forecast
n_periods = len(test)
sarima_forecast = sarima_result.forecast(steps=n_periods)

```

```python
# ‚úÖ Plot Actual vs Forecast
plt.figure(figsize=(12,6))
plt.plot(train.index, train['Close'], label='Train', color='blue')
plt.plot(test.index, test['Close'], label='Test', color='green')
plt.plot(test.index, sarima_forecast, label='SARIMA Forecast', color='red')
plt.title('SARIMA Model Forecast')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

# LSTM MODEL

```python
## Scaling and sequence prep
# Scale data (0‚Äì1 range)
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

train_scaled = scaled_data[:len(train)]
test_scaled = scaled_data[len(train):]

# Create sequences (60 days ‚Üí next day prediction)
def create_sequences(data, seq_length=60):
    X, y = [], []
    for i in range(seq_length, len(data)):
        X.append(data[i-seq_length:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train_scaled)
X_test, y_test = create_sequences(test_scaled)

X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
## Build and train LSTM model

model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

```python
## Predictions

predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions.reshape(-1,1))

true_prices = df.iloc[len(df)-len(predictions):]['Close']

rmse_lstm = sqrt(mean_squared_error(true_prices, predictions))
mae_lstm = mean_absolute_error(true_prices, predictions)

print("LSTM RMSE:", rmse_lstm)
print("LSTM MAE:", mae_lstm)

plt.figure(figsize=(10,5))
plt.plot(true_prices.index, true_prices, label='Actual Price')
plt.plot(true_prices.index, predictions, label='LSTM Prediction')
plt.legend()
plt.title("LSTM Forecast vs Actual")
plt.show()
```

```python
#@title Result Comparison

results = pd.DataFrame({
    'Model': ['Naive', 'ARIMA', 'LSTM'],
    'RMSE': [rmse_naive, rmse_arima,rmse_lstm],
    'MAE': [mae_naive, mae_arima, mae_lstm]
})
print(results)
```
